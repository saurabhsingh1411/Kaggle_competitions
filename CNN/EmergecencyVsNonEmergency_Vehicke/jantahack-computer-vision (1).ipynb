```python
train_dir='../input/janata-hack-image-classification/images/images'
train_fnames=os.listdir(train_dir)
print(train_fnames[0:10])
print('total training images :', len(os.listdir(train_dir) ))
```

```python
train_imgname=pd.read_csv('../input/janata-hack-image-classification/train.csv')
train_imgname.head()
```

```python
imgname=train_imgname['image_names']
imgname=list(np.array(imgname))
imgname[5:10]
```

```python
import cv2
X=[]
for ele in imgname:
    img=cv2.imread('../input/janata-hack-image-classification/images/images/'+ele)
    X.append(img)
X=np.array(X)

print(X.shape)
```

```python
X=X/255
print(X.shape)
```

```python
y=train_imgname['emergency_or_not']
y.tail()
```

Creating Model


```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from tqdm import tqdm
import tensorflow as tf
```

```python
model=tf.keras.models.Sequential()

model.add(Conv2D(16, (3,3), activation='relu', input_shape=(224, 224, 3)))
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.2))
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.2))
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
from tensorflow.keras.optimizers import RMSprop

model.compile(optimizer=RMSprop(lr=0.001),
              loss=keras.losses.binary_crossentropy,
              metrics=['accuracy'])
```

```python
history = model.fit(X,y,steps_per_epoch=100,epochs=5,validation_split=0.2,verbose=2)
```
